```python
import pandas as pd
import numpy as np

data_path = "../data/raw/creditcard.csv"
df = pd.read_csv(data_path)
```

```python
labels = df["Class"]
df_features = df.drop(columns=["Class"])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_features["Amount_scaled"] = scaler.fit_transform(
    df_features[["Amount"]]
)
```

- Large amounts don’t dominate

- Models focus on behavior, not magnitude

```python
df_features["transaction_count"] = 1

df_features["tx_per_min"] = (
    df_features["transaction_count"]
    .rolling(window=60)
    .sum()
)
```

**How fast are transactions happening?**
- Each row = 1 transaction

- Rolling window = 60 seconds

- High values = rapid activity → suspicious

```python
df_features["rolling_mean_amount"] = (
    df_features["Amount"]
    .rolling(window=60)
    .mean()
)

df_features["rolling_std_amount"] = (
    df_features["Amount"]
    .rolling(window=60)
    .std()
)
```

- Sudden spikes in amount = red flag

- Volatility matters more than raw value

```python
df_features["hour"] = (df_features["Time"] // 3600) % 24
```

- Convert seconds → hour of day

- Some fraud clusters at odd hours

```python
df_features = df_features.fillna(0)
```
